## 重点分析
- tradeMoney：真是租金值
- pred_tradeMoney：预测租金值

通过这两项我们分析下我们预测误差出现错误的原因

```python
true_pred['error']=abs(true_pred['tradeMoney']-true_pred['pred_tradeMoney'])
```

### 分析误差

- 误差大于10000

```python
true_pred[true_pred['error']>=10000]
```

可以发现，一些面积大的房屋真实租金很低，但是我们预测租金很大，同时这也和房屋所在地域密切相关

- 误差大于4000小于10000

```python
true_pred.query("4000<=error<10000").head()
```

```python
len(true_pred.query("4000<=error<10000"))
```

- 将误差从大到小排序

```python
true_pred.sort_values(ascending=False,by="error")[:30]
```

```python
true_pred['error'].describe()
```

### 分析房屋面积

```python
true_pred.sort_values(ascending=False,by="area")[:500]
```

```python
true_pred.sort_values(ascending=False,by="area")[-30:]
```

```python
true_pred['area'].describe()
```

- 通过训练模型的得到特征重要性可知，房屋面积area起着非常重要的作用，基本上可以认为房屋面积越大，租金越高；但是并不是所有的情况，还需结合房屋其他情况进行结合分析

- 从上面的房屋面积排序结果来看，一些房屋面积存在一些异常值，我们需要加以处理
